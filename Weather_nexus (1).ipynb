```python
data['Sunshine'].fillna(0,inplace=True)
data['WindGustDir'].fillna(0,inplace=True)
data['WindGustSpeed'].fillna(0,inplace=True)
data['WindDir9am'].fillna(0,inplace=True)
data['WindDir3pm'].fillna(0,inplace=True)
data['WindSpeed9am'].fillna(0,inplace=True)
```

```python
basic_stats = data[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation']].describe()
```

```python
#Basic Statistics
basic_stats
```

```python
#Timeseries Visualization
columns = ['MinTemp', 'MaxTemp', 'Rainfall']
data['index'] = range(len(data))
plt.figure(figsize=(15, 6))
for column in columns:
    plt.plot(data['index'], data[column], label=column)
plt.title('Time Series Visualization')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#correlation Analysis
correlation_matrix = data[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Analysis')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.hist(data['Rainfall'], bins=30, edgecolor='black')
plt.title('Rainfall Distribution')
plt.xlabel('Rainfall')
plt.ylabel('Frequency')
plt.show()
```

```python

data['Month'] = (data.index % 12) + 1

# Convert numeric columns to numeric type, ignoring errors for non-numeric values
numeric_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                   'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
                   'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
seasonal_averages = data.groupby('Month')[numeric_columns].mean()

plt.figure(figsize=(18, 10))

for i, column in enumerate(numeric_columns):
    plt.subplot(4, 4, i + 1)
    plt.bar(seasonal_averages.index, seasonal_averages[column])
    plt.title(f'{column} - Seasonal Patterns')
    plt.xlabel('Month')
    plt.ylabel('Average Value')

plt.tight_layout()
plt.show()
```
